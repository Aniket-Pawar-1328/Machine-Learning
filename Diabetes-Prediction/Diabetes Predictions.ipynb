##  Feature Engineering / Data preprocessing 

### Feature Encoding

```python
health_data['diabetes'].value_counts()
```

Using label encoding to encode the "gender" and "diabetes" column

```python
health_data['gender'].value_counts()
```

```python
def gender_encode(x):
    if x == 'female':
        return 0
    else:
        return 1
health_data['gender'] = health_data['gender'].apply(lambda x : gender_encode(x))
```

```python
health_data['gender'].value_counts()
```

```python
def diabetes_encode(x):
    if x == 'No diabetes':
        return 0
    else:
        return 1
health_data['diabetes'] = health_data['diabetes'].apply(lambda x : diabetes_encode(x))
```

```python
health_data['diabetes'].value_counts()
```

```python
health_data.head()
```

### Handling Imbalance

The model is highly imbalanced so randomly sampling the data for further model buildup

```python
health_data['diabetes'].value_counts()
```

```python
diabetes_yes = health_data[health_data['diabetes']==1]
diabetes_yes
```

```python
diabetes_no = health_data[health_data['diabetes']==0].sample(60)
diabetes_no
```

```python
health_data1 = pd.concat([diabetes_yes, diabetes_no], axis=0).sample(frac=1, random_state=42).reset_index(drop=True)
health_data1
```

```python
health_data1['diabetes'].value_counts()
```

### Seperating the Target and independent variables 

```python
x = health_data1.drop('diabetes',axis=1)
x
```

```python
y = health_data1['diabetes']
y
```

### Splitting Testing and Training Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.value_counts()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
num_col
```

```python
x_train[num_col]
```

```python
x_train[num_col] = scaler.fit_transform(x_train[num_col])
x_train
```

```python
x_test[num_col] = scaler.transform(x_test[num_col])
x_test
```

## Model Development

### Importing Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
lr_model = LogisticRegression()
dt_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
knn_model = KNeighborsClassifier()
svm_model = SVC()
nb_model = GaussianNB()
```

```python
models = [lr_model,dt_model,rf_model,knn_model,svm_model,nb_model]
models
```

### Model Training

```python
for model in models:
    model.fit(x_train,y_train)
```

### Model Evaluation

#### Classification Report

```python
from sklearn.metrics import classification_report
```

```python
for model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(model.__class__.__name__)
    print(classification_report(y_test,y_pred))
    print("-------------------------------------------------------")
    print(" ")
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
for model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(model.__class__.__name__)
    cm = confusion_matrix(y_test,y_pred)
    sns.heatmap(cm, annot=True, fmt= " ")
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    print("-------------------------------------------------------")
    print(" ")
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
%%time
for model in models:
    acc_ls =  cross_val_score(model,x,y, scoring='accuracy', cv=5)
    mean_acc = round(acc_ls.mean()*100,2)
    modelname = model.__class__.__name__
    print(f"{modelname}'s Cross Validation Accuracy is = {mean_acc}%")
    print("-------------------------------------------------------")
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30, None],
    'criterion': ['gini', 'entropy']
}
```

```python
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
%%time
grid_search.fit(x_train, y_train)
```

```python
# Best hyperparameters
print("Best hyperparameters found: ", grid_search.best_params_)
```

```python
best_estimator = grid_search.best_estimator_
best_estimator.fit(x_train,y_train)
y_pred_best = best_estimator.predict(x_test)
print(classification_report(y_test,y_pred_best))
```

### Customizing predictions

```python
def predict_diabetes(model, scaler, num_col):
    print("Enter Patient's Data:")
    print(" ")
    cholesterol_in = int(input("Enter Cholesterol Level = "))
    glucose_in = int(input("Enter Glucose Level = "))
    hdl_in = int(input("Enter HDL Level = "))
    chol_hdl_ratio_in = (cholesterol_in/hdl_in)
    age_in = int(input("Enter Patient's Age = "))
    
    gender_in = input("Enter Patient's Gender (0 for Female, 1 for Male)= ")
    while gender_in not in ['0', '1']:
        print("Invalid input. Please enter 0 or 1")
        gender_in = input("Enter the Gender: ")
    gender_in = int(gender_in)
    
    height_in = int(input("Enter Height = "))
    weight_in = int(input("Enter Weight = "))
    bmi_in = (weight_in * 703 /(height_in * height_in))
    systolic_bp_in = int(input("Enter systolic bp = "))
    diastolic_bp_in = int(input("Enter diastolic bp = "))
    waist_in = int(input("Enter waist = "))
    hip_in = int(input("Enter hip = "))
    waist_hip_ratio_in = (waist_in/hip_in)

    input_data = {
        "glucose": [glucose_in],
        "chol_hdl_ratio": [chol_hdl_ratio_in],
        "age": [age_in],
        "gender": [gender_in],
        "bmi": [bmi_in],
        "systolic_bp": [systolic_bp_in],
        "diastolic_bp": [diastolic_bp_in],
        "waist_hip_ratio": [waist_hip_ratio_in]
    }
    
    input_df = pd.DataFrame(input_data)
    input_df[num_col] = scaler.transform(input_df[num_col])    
    input_for_prediction = input_df.values.reshape(1, -1)
    prediction = model.predict(input_for_prediction)
    
    if prediction[0] == 1:
        return "The patient has diabetes."
    else:
        return "The patient does not have diabetes."
```
